```python
df['START_DATE*']=df['START_DATE*'].astype('datetime64[ns]')
df['END_DATE*']=df['END_DATE*'].astype('datetime64[ns]')
```

```python
df['Time'] = df['END_DATE*']-df['START_DATE*'] #find time required by substracting the two values
df.head()
```

```python
df['Time'] = df['Time'].dt.total_seconds()/60.0
df['Time'] = df['Time'].astype('float')
df.head()
```

```python
df['KM']= df['MILES*']*1.609 #converted miles to km for better understanding
```

```python
df.head()
```

```python
df.drop(['MILES*'],axis=1) #dropped miles since km is present
```

```python
df['speed']=(df['KM']/df['Time'])*60 #time:min & speed:kmph
```

```python
#HOUR DAY TRANSFORMATION
```

```python
import datetime
import calendar
```

```python
df['START_DATE*'] = pd.to_datetime(df['START_DATE*'], format="%m/%d/%Y %H:%M") 
df['END_DATE*'] = pd.to_datetime(df['END_DATE*'], format="%m/%d/%Y %H:%M")
```

```python
hour=[]  #empty list
day=[]
dayofweek=[]
month=[]
weekday=[]
for x in df['START_DATE*']:
    hour.append(x.hour) #adding/appending the values to above empty list
    day.append(x.day)
    dayofweek.append(x.dayofweek)
    month.append(x.month)
    weekday.append(calendar.day_name[dayofweek[-1]])
df['HOUR']=hour #creatig columns
df['DAY']=day
df['DAY_OF_WEEK']=dayofweek
df['MONTH']=month
df['WEEKDAY']=weekday
```

```python
df.head() #updated dataset
```

```python
plt.figure(figsize=(20,20))
df['HOUR'].value_counts().plot(kind='bar',figsize=(10,5),color='red') #trips per hour
plt.show()
```

```python
plt.figure(figsize=(16,9))
df['WEEKDAY'].value_counts().plot(kind='bar',figsize=(10,5),color='black') #trips per day of week
plt.show()
```

```python
df['DAY'].value_counts().plot(kind='bar',figsize=(10,5),color='blue') #trips per day of the month
plt.show()
```

```python
plt.figure(figsize=(20,20))
df['MONTH'].value_counts().plot(kind='bar',figsize=(10,5),color='green') #trips in month, month 12 has highest number of trips as depicted in the figure
plt.show()
```
